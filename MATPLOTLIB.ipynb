## Set Font Properties for Title and Labels

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints=np.array([88,86,100,97])
xpoints=np.array(["English","VDC","MECH","PHY"])
font1 = {'family':'serif','color':'k','size':20}
font2 = {'family':'serif','color':'green','size':15}

plt.plot(xpoints,ypoints)
plt.title("Semester 1 results",fontdict=font1)
plt.xlabel("Subjects",fontdict=font2)
plt.ylabel("Marks",fontdict=font2)
plt.show
```

## Subplot with title and labels
